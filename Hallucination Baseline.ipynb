```python
import boto3

bedrock_agent = boto3.client('bedrock-agent-runtime', region_name='us-east-1')
KB_ID = "8OOQBDOPXT"

queries = [
    "Who is Sophia Chen and where does she live?",
    "What is Fuquay-Varina known for?",
    "What is Dr. Raj Patel's profession?"
]

for query in queries:
    print(f"\n=== Query: {query} ===")
    response = bedrock_agent.retrieve(
        knowledgeBaseId=KB_ID,
        retrievalQuery={"text": query},
        retrievalConfiguration={
            "vectorSearchConfiguration": {
                "numberOfResults": 3
            }
        }
    )
    for i, ref in enumerate(response.get('retrievalResults', []), 1):
        text = ref.get('content', {}).get('text', 'No text')
        clean_text = text.replace('\n', ' ').strip()[:200]
        print(f"Chunk {i}: {clean_text}...\n")
```
